```python
import numpy as np
import pygame
from minesweeper import Minesweeper
```

```python
num_mine = 50
sizeX = 16
sizeY = 16


### pygame 전역변수 선언 ###

# colors
BLACK  = (0, 0, 0)
WHITE  = (255, 255, 255)
RED    = (255, 0, 0)
GREEN  = (0, 255, 0)
BLUE   = (0, 0, 255)
GRAY64 = (64, 64, 64)

# screen
size   = [1280, 720]

# 시스템 관리
state = 0 # 0: 로비, 1: 설정, 2: 게임
done = False

# figure
ONE   = pygame.image.load('./assets/figure/one.jpg')
ONE   = pygame.transform.scale(ONE, (30,30))
TWO   = pygame.image.load('./assets/figure/two.jpg')
TWO   = pygame.transform.scale(TWO, (30,30))
THREE = pygame.image.load('./assets/figure/three.jpg')
THREE = pygame.transform.scale(THREE, (30,30))
FOUR  = pygame.image.load('./assets/figure/four.jpg')
FOUR  = pygame.transform.scale(FOUR, (30,30))
FIVE  = pygame.image.load('./assets/figure/five.jpg')
FIVE  = pygame.transform.scale(FIVE, (30,30))
SIX   = pygame.image.load('./assets/figure/six.jpg')
SIX   = pygame.transform.scale(SIX, (30,30))
SEVEN = pygame.image.load('./assets/figure/seven.jpg')
SEVEN = pygame.transform.scale(SEVEN, (30,30))
EIGHT = pygame.image.load('./assets/figure/eight.jpg')
EIGHT = pygame.transform.scale(EIGHT, (30,30))
MINE  = pygame.image.load('./assets/figure/mine.jpg')
MINE  = pygame.transform.scale(MINE, (30,30))
STAR  = pygame.image.load('./assets/figure/star.jpg')
STAR  = pygame.transform.scale(STAR, (30,30))
```

```python
pygame.init() # pygame 초기화

game = Minesweeper(num_mine=50, sizeX=sizeX, sizeY=sizeY)
#print('mines')
#print(game.mines)
#print('num_mine_around')
#print(game.num_mine_around)
screen = pygame.display.set_mode(size)
clock = pygame.time.Clock()

# BGM
#bgm = pygame.mixer.Sound('./assets/bgm/03_Game_Theme.mp3')
#bgm.play(-1) # -1 시 무한반복

# font
font_title  = pygame.font.SysFont('arial', 80, True, True) # (font name, size, bold=False, italic=False)
title_lobby = font_title.render('MINESWEEPER', True, BLACK) # (Text, antialias, color, background=None)

new_game_font = font_title.render('NEW', True, BLACK)
new_game_button_xywh = (940, 310, 200, 100)
### pygame 루프 ###
def runGame():
    cell_size  = 30
    num_X = sizeX
    num_Y = sizeY
    grid_offset = [400, 150]

    global done
    while not done:
        clock.tick(20) # clock, tick으로 FPS(Frame per Second) 개념
        screen.fill(WHITE)

        state = game.get_state()

        ### 유저 입력 감지 및 처리 ###
        for event in pygame.event.get():
            if event.type == pygame.QUIT: # 종료
                done = True
                
            if event.type == pygame.MOUSEBUTTONDOWN:
                if event.button == 1: # 1: left click / 좌클릭:open
                    
                    if (event.pos[0] > 940) and (event.pos[0] < 1140) and (event.pos[1] > 310) and (event.pos[1] < 410): # set 완료
                        game.new_game(num_mine=num_mine, sizeX=sizeX, sizeY=sizeY)
                    else:
                        posX = (event.pos[0] - grid_offset[0]) // cell_size
                        posY = (event.pos[1] - grid_offset[1]) // cell_size
                        num = game.pos2num(posX, posY)
                        if not (posX < 0) and not (posX > 15) and not (posY < 0) and not (posY > 15):
                            game.action(num)
                
                elif event.button == 3: # 3: right click / 우클릭:point
                    posX = (event.pos[0] - grid_offset[0]) // cell_size
                    posY = (event.pos[1] - grid_offset[1]) // cell_size
                    num = game.pos2num(posX, posY) + game.num_cell
                    if not (posX < 0) and not (posX > 15) and not (posY < 0) and not (posY > 15):
                        game.action(num)


        ### 화면 표시 ###       

        # New game
        screen.blit(new_game_font, (940, 310))
        pygame.draw.rect(screen, BLACK, new_game_button_xywh, 1)

        # grid
        for posX in range(num_X):
            for posY in range(num_Y):
                if state[posY][posX] == -5: # Unknown cell: GRAY64
                    rect = (grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY, cell_size, cell_size)
                    pygame.draw.rect(screen, GRAY64, rect, 0) # 1은 width인데, 0 이상이면 color의 테두리 rect / 0이면 테두리 없이 color로 fill된 rect
                elif state[posY][posX] == -1: # pointed cell: 
                    screen.blit(STAR, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])

                if not (state[posY][posX] == -5) and not (state[posY][posX] == -1) and game.mines[posY][posX] == 1: # opened mine cell
                    screen.blit(MINE, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])
                elif state[posY][posX] == 0: # mine cell이 아니고 num_mine_around가 0이면 WHITE
                    rect = (grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY, cell_size, cell_size)
                    pygame.draw.rect(screen, WHITE, rect, 0)
                elif state[posY][posX] == 1:
                    screen.blit(ONE, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])
                elif state[posY][posX] == 2:
                    screen.blit(TWO, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])
                elif state[posY][posX] == 3:
                    screen.blit(THREE, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])
                elif state[posY][posX] == 4:
                    screen.blit(FOUR, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])
                elif state[posY][posX] == 5:
                    screen.blit(FIVE, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])
                elif state[posY][posX] == 6:
                    screen.blit(SIX, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])
                elif state[posY][posX] == 7:
                    screen.blit(SEVEN, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])
                elif state[posY][posX] == 8:
                    screen.blit(EIGHT, [grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY])

                rect = (grid_offset[0] + cell_size * posX, grid_offset[1] + cell_size * posY, cell_size, cell_size)
                pygame.draw.rect(screen, BLACK, rect, 1) # 1은 width인데, 0 이상이면 color의 테두리 rect / 0이면 테두리 없이 color로 fill된 rect


        pygame.display.update()
runGame()
pygame.quit()
```
